```python
from scipy.stats import norm
from scipy.integrate import quad

# تابع چگالی احتمال نرمال استاندارد
f = norm.pdf

# انتگرال PDF بین -4 تا 4
area, _ = quad(f, -4, 4)

# احتمال بیرون از بازه
p_outside = 1 - area
print(p_outside)
```
