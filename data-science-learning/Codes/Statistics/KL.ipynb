```python
import numpy as np
from scipy.stats import norm
from scipy.stats import poisson
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
p = [0.10, 0.41, 0, 0.334] #احتمال اینکه داده ها در اون بین بیوفتن (پراکندگی)
q = [0.11, 0.45, 0.12, 0.345]
```

```python
def kl(p, q): # جای این دوتا عوض شه عدد هم عوض میشه
    result = np.sum(np.where(np.logical_and(p != 0, q != 0), p * np.log(p / q), 0))
    return result
```

```python
x = np.arange(-10, 10, 0.001)
len(x)
```

```python
p = norm.pdf(x, 0, 2)
q = norm.pdf(x, 2, 3)
plt.plot(x, p, c='blue')
plt.plot(x, q, c='red')
```

```python
kl(p, q)
```

```python
p = norm.pdf(x, 0, 2)
q = norm.pdf(x, 0.1, 2.5)
plt.plot(x, p, c='blue')
plt.plot(x, q, c='red')
```

```python
kl(p, q)
```

```python
import numpy as np

np.random.seed(0)
data_p = np.random.normal(0, 1, 1000)   # توزیع نرمال با میانگین 0 و انحراف معیار 1
data_q = np.random.normal(0.5, 1, 1000) # توزیع نرمال با میانگین 0.5 و انحراف معیار 1
```

```python
bins = 30  # تعداد بازه‌ها (bins)

p_hist, bin_edges = np.histogram(data_p, bins=bins, density=True)
q_hist, _ = np.histogram(data_q, bins=bin_edges, density=True)   #هر مقدار در این آرایه‌ها معادل یک bin از هیستوگرام است
```

```python
#نرمال سازی تا مجموع =1 شود
p = p_hist / np.sum(p_hist)
q = q_hist / np.sum(q_hist)
```
