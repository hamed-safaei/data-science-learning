```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ساخت داده نمونه (شبیه df['comb08'])
np.random.seed(0)
df = pd.DataFrame({'comb08': np.random.normal(25, 5, 500)})

# میانگین داده‌ها
mean_value = df['comb08'].mean()

# ساخت شکل و محور
fig, ax = plt.subplots(figsize=(8, 5))

# رسم هیستوگرام
ax.hist(df['comb08'], bins=20, color='black', alpha=0.5)

# اضافه کردن خط میانگین
ax.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_value:.2f}')

# تنظیمات ظاهری
ax.set_title('Distribution of comb08', fontsize=14)
ax.set_xlabel('comb08', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.legend()

# نمایش نمودار
plt.show()
```

```python
# ===== داده نمونه =====
np.random.seed(0)
df = pd.DataFrame({
    'comb08': np.random.normal(25, 5, 500),
    'comb10': np.random.normal(30, 7, 500)
})

# ===== میانگین‌ها =====
mean_comb08 = df['comb08'].mean()
mean_comb10 = df['comb10'].mean()

# ===== ساخت شکل و محورهای چندگانه =====
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # 2 ردیف، 2 ستون
fig.suptitle('Comprehensive Example of Subplots', fontsize=16)  # عنوان کلی شکل
fig.subplots_adjust(hspace=0.4, wspace=0.3)  # فاصله بین نمودارها

# ===== نمودار 1: هیستوگرام comb08 =====
axs[0, 0].hist(df['comb08'], bins=20, color='skyblue', alpha=0.7)
axs[0, 0].axvline(mean_comb08, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_comb08:.2f}')
axs[0, 0].set_title('Histogram of comb08')
axs[0, 0].set_xlabel('comb08')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].grid(True)
axs[0, 0].legend()

# ===== نمودار 2: هیستوگرام comb10 =====
axs[0, 1].hist(df['comb10'], bins=15, color='lightgreen', alpha=0.6)
axs[0, 1].axvline(mean_comb10, color='orange', linestyle='-.', linewidth=2, label=f'Mean = {mean_comb10:.2f}')
axs[0, 1].set_title('Histogram of comb10')
axs[0, 1].set_xlabel('comb10')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].grid(True)
axs[0, 1].legend()

# ===== نمودار 3: نمودار خطی comb08 =====
axs[1, 0].plot(df['comb08'], color='purple', linestyle='-', marker='o', markersize=4, alpha=0.7, label='comb08 values')
axs[1, 0].axhline(mean_comb08, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_comb08:.2f}')
axs[1, 0].set_title('Line Plot of comb08')
axs[1, 0].set_xlabel('Index')
axs[1, 0].set_ylabel('Value')
axs[1, 0].grid(True)
axs[1, 0].legend()

# ===== نمودار 4: نمودار پراکندگی comb08 vs comb10 =====
axs[1, 1].scatter(df['comb08'], df['comb10'], color='darkcyan', alpha=0.6)
axs[1, 1].set_title('Scatter Plot comb08 vs comb10')
axs[1, 1].set_xlabel('comb08')
axs[1, 1].set_ylabel('comb10')
axs[1, 1].grid(True)

# ===== ذخیره شکل =====
fig.savefig('comprehensive_example.png', dpi=300)

# ===== نمایش شکل =====
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# داده نمونه
data = np.random.normal(25, 5, 1000)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(data, bins=10, range=(10,50), color='skyblue', alpha=0.7)

# patches[2].get_x() 


# اضافه کردن مقدار روی هر ستون
for rect, value in zip(patches, n):
    height = rect.get_height()        # ارتفاع ستون
    x = rect.get_x() + rect.get_width()/2  # مرکز ستون
    ax.text(x, height + 1, f'{int(value)}', ha='center', va ='bottom')  # قرار دادن متن
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ساخت داده‌ها
np.random.seed(42)
data1 = np.random.normal(25, 5, 100)
data2 = np.random.normal(30, 7, 100)
data3 = np.random.normal(22, 4, 100)

# ساخت شکل و محور
fig, ax = plt.subplots(figsize=(8, 6))

# رسم نمودار جعبه‌ای
box = ax.boxplot(
    [data1, data2, data3],
    patch_artist=True,      # برای رنگی کردن جعبه‌ها
    labels=['Set A', 'Set B', 'Set C'],
    showmeans=True,         # نمایش میانگین با علامت مثلث
    meanline=False,         # نمایش میانگین به صورت نقطه‌ای
    notch=True              # تورفتگی در اطراف میانه (برای نشان دادن فاصله اطمینان)
)

# رنگ دادن به هر جعبه
colors = ['#66b3ff', '#99ff99', '#ffcc99']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# اضافه کردن عنوان و برچسب‌ها
ax.set_title('Advanced Box Plot Example', fontsize=16)
ax.set_ylabel('Values', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# نوشتن مقدار میانه روی هر جعبه
for line, label in zip(box['medians'], ['Set A', 'Set B', 'Set C']):
    x, y = line.get_xydata()[1]
    ax.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```
